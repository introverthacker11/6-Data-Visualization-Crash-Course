###
## Secondary Axis
###

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

a = sns.load_dataset('iris')
print(a.head(7))

x = a[['sepal_width','sepal_length']]
y = a[['petal_width','petal_length']]
plt.rcParams['figure.figsize'] = [8,4]

ax = x.plot()
ax2 = y.plot(ax = ax, secondary_y = True)

ax.set_ylabel('Sepal (Width and Length)\n')
ax.set_xlabel('\nIndex Numbers (0-150)')
ax2.set_ylabel('\nPetal (Width and Length)')
ax.set_title('Plot using Secondary Axis\n')
```
