###
## boxplot
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot( kind = 'box')
plt.show()
```

###
## boxplot properties
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot( kind = 'box', color = 'red', whiskerprops = { 'color' : 'blue'}, medianprops = { 'color' : 'white'}, capprops = {'color' : 'yellow'}, patch_artist = True)
plt.show()
```

###
## boxplot horizontal
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot( kind = 'box', color = 'red', whiskerprops = { 'color' : 'blue'}, medianprops = { 'color' : 'white'}, capprops = {'color' : 'yellow'}, patch_artist = True, vert = False)
plt.show()
```
