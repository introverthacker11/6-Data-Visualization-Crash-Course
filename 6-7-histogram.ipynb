###
## hist
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot.hist()
a.plot(kind = 'hist')
```

###
## bins 
It tell the number of bars or segmentations

###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot(kind = 'hist', bins = 7)
```

###
## stacked histogram plot
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot(kind = 'hist', bins = 7, stacked = True)
```

###
## orientation
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a.plot(kind = 'hist', bins = 7, stacked = True, orientation = 'horizontal', figsize = (8,4))
```

###
## petal_length
###

```python

import seaborn as sns
import matplotlib.pyplot as plt

a = sns.load_dataset('iris')
a['petal_length'].plot(kind = 'hist', bins = 7, stacked = True, orientation = 'horizontal', figsize = (8,4))
```
