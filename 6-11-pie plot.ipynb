###
## pie chart
###

```python

import matplotlib.pyplot as plt
import seaborn as sns

a = sns.load_dataset('iris').drop('species', axis = 1)
a.head()
```

```python
a.iloc[0].plot.pie(figsize = (5,5), fontsize = 10, autopct = '%1.1f%%')
```

###
## sub pie plots
###

```python

import matplotlib.pyplot as plt
import seaborn as sns

a = sns.load_dataset('iris').drop('species', axis = 1)
a.head(3).T.plot.pie(subplots = True, legend = False, fontsize = 10, figsize = (10,10), autopct = '%1.1f%%')
#plt.show()
plt.tight_layout()
```

###
## legends sub pie plots
###

```python

import matplotlib.pyplot as plt
import seaborn as sns

a = sns.load_dataset('iris').drop('species', axis = 1)
ax = a.head(3).T.plot.pie(subplots = True, fontsize = 8, figsize = (10,10), autopct = '%1.1f%%', legend = True)

# changing legend location
ax[0].legend(bbox_to_anchor = (1,1.5))
ax[1].legend(bbox_to_anchor = (1,1.5))
ax[2].legend(bbox_to_anchor = (1,1.5))
plt.tight_layout()
plt.show()
```
